```python
import matplotlib.pyplot as plt
import numpy as np
from contpy import optimize
```

```python
%%time
a1, b1 = 0, 0.5
w = 2
xs = lambda p : (a1 + b1*p)*np.cos(w*p)
ys = lambda p : (a1 + b1*p)*np.sin(w*p)               
spiral_res_vec = lambda x, p : np.array([(x[0] - (a1 + b1*p)*np.cos(w*p)), (x[1] - (a1 + b1*p)*np.sin(w*p))]) 

x0=np.array([0.0,0.0])
p_range=(0.0,10.0)
x_sol, p_sol, info_dict = optimize.continuation(spiral_res_vec,x0=x0,p_range=p_range,p0=0.0,step=0.5,max_int=500)
```

```python
# target curve
p_list = np.linspace(p_range[0],p_range[1],100)
xs_target = list(map(xs,p_list))
ys_target = list(map(ys,p_list))

# ploting continuation curve
plt.figure()
plt.plot(x_sol[0,:],x_sol[1,:],'k.', label='Continuation points')
plt.plot(xs_target,ys_target,'r-', label='Target curve')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
```
